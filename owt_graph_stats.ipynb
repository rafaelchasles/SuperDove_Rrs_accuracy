```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo os dados do arquivo CSV
df = pd.read_csv('/content/drive/MyDrive/mestrado/dissertacao/dados/tabelas/resultados/erros_owt_graph.csv', sep=';')

# Definindo os valores de x
valores_x = df.iloc[:, 0]

# Criando os gráficos para cada métrica
plt.figure(figsize=(10, 12))

# Plotando e_DSF com e_PSR
plt.subplot(3, 2, 1)
plt.plot(valores_x, df['e'], label='ε', marker='o', linestyle='--',markersize=3, linewidth=1)
plt.xlabel('OWT', fontsize=8)
plt.ylabel('ε (%)', fontsize=8)
plt.title('ε (%)', fontsize=10)
plt.ylim(0, 160)  # Limitando o eixo y para 0 a 100
plt.legend(fontsize=8)
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(valores_x, df['b'], label='β', marker='o', linestyle='--',markersize=3, linewidth=1)
plt.xlabel('OWT', fontsize=8)
plt.ylabel('β (%)', fontsize=8)
plt.title('β (%)', fontsize=10)
plt.ylim(0, 160)  # Limitando o eixo y para 0 a 100
plt.legend(fontsize=8)
plt.grid(True)

plt.subplot(3, 2, 3)
plt.plot(valores_x, df['MAPE'], label='MAPE', marker='o', linestyle='--',markersize=3, linewidth=1)
plt.xlabel('OWT', fontsize=8)
plt.ylabel('MAPE', fontsize=8)
plt.title('MAPE', fontsize=10)
plt.ylim(0, 500)  # Limitando o eixo y para 0 a 100
plt.legend(fontsize=8)
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(valores_x, df['RMSLE'], label='RMSLE', marker='o', linestyle='--',markersize=3, linewidth=1)
plt.xlabel('OWT', fontsize=8)
plt.ylabel('RMSLE', fontsize=8)
plt.title('RMSLE', fontsize=10)
plt.ylim(0, 0.03)  # Limitando o eixo y para 0 a 100
plt.legend(fontsize=8)
plt.grid(True)

plt.subplot(3, 2, 5)
plt.plot(valores_x, df['s'], label='Slope', marker='o', linestyle='--',markersize=3, linewidth=1)
plt.xlabel('OWT', fontsize=8)
plt.ylabel('Slope', fontsize=8)
plt.title('Slope', fontsize=10)
plt.ylim(0, 0.7)  # Limitando o eixo y para 0 a 100
plt.legend(fontsize=8)
plt.grid(True)

plt.tight_layout()
plt.show()
```
