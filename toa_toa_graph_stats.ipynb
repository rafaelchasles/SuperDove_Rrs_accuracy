```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo os dados do arquivo CSV
df = pd.read_csv('/content/drive/MyDrive/mestrado/dissertacao/dados/tabelas/resultados/toa_toa_stats.csv', sep=';', encoding='latin1')

# Definindo os valores de x
valores_x = df.iloc[:, 0]

# Criando os gráficos para cada métrica
plt.figure(figsize=(15, 10))

# Plotando e
plt.subplot(2, 2, 1)
bar_width = 0.35
index = range(len(valores_x))

# Gráfico de barras para 'e'
plt.bar(index, df['e'], width=bar_width, label='ε (%)', color='blue', alpha=0.7)

# Gráfico de barras para 'b'
plt.bar([p + bar_width for p in index], df['b'], width=bar_width, label='β (%)', color='orange', alpha=0.7)

plt.ylabel('(%)')
plt.title('ε e β TOA MSI vs. TOA PSD')
plt.ylim(-22, 40)  # Limitando o eixo y
plt.legend()
plt.grid(True, axis='y')

plt.axhline(0, color='black', linewidth=2)

# Ajustando os ticks no eixo x
plt.xticks([p + bar_width/2 for p in index], valores_x, rotation=25)

plt.tight_layout()
plt.show()
```
