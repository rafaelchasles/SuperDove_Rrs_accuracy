```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
```

```python
df = pd.read_excel("/content/drive/MyDrive/mestrado/dissertacao/dados/tabelas/resultados/espectros_insitu3.xlsx", index_col='id')
```

```python
df = df.transpose().set_index(pd.Index(range(400,901))).transpose()
```

```python
df
```

```python
fig, ax = plt.subplots()

for i in range(len(df)):

  ax.plot(df.columns, df.iloc[i])

# Personalizar gráfico
#ax.set_xlabel('')
#ax.set_ylabel('')
#ax.legend(loc='upper left')
ax.set_xlabel('Comprimento de Onda')
ax.set_ylabel('Rrs')
# Mostrar gráfico
plt.show()
```

```python
# Realize o groupby por 'idd'
grupos = df.groupby(level='id')

# Iterar sobre os grupos e plotar os gráficos
for idd, grupo in grupos:
    fig, ax = plt.subplots()
    for i in range(len(grupo)):
        ax.plot(grupo.columns, grupo.iloc[i], color='black' , linewidth=0.6)
    ax.set_xlabel('Comprimento de Onda (nm)')
    ax.set_ylabel('Rrs (sr-¹)')
    ax.set_title( f'{idd}')  # Adicione um título com o idd do grupo
    plt.show()
```
