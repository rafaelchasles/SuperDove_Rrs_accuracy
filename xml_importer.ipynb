```python
import pandas as pd
import os
from lxml import etree
from bs4 import BeautifulSoup

# ler arquivo CSV
df = pd.read_csv('D:/mestrado/dissertacao/dados/tabelas/resultados/Amz1/DSF_glint_stats.csv')

# diretório que contém as pastas com os arquivos XML
xml_dir = 'E:/rasters/Amz/TOA'

# iterar sobre as linhas do dataframe
for index, row in df.iterrows():
    img_id = row['img_id']
    img_dir = os.path.join(xml_dir, img_id)

    # verifique se a pasta existe antes de continuar
    if os.path.exists(img_dir):
        # listar todos os arquivos na pasta correspondente a imagem
        img_files = os.listdir(img_dir)

        # encontrar o arquivo XML correspondente
        xml_file = None
        for f in img_files:
            if f.endswith('.xml'):
                xml_file = os.path.join(img_dir, f)
                break

        # verifique se o arquivo existe antes de abri-lo
        if xml_file is not None and os.path.exists(xml_file):
            with open(xml_file, 'r') as f:
                xml_content = f.read()

            # parseie o conteúdo XML com BeautifulSoup
            soup = BeautifulSoup(xml_content, 'xml')

            # encontre o valor de <eop:serialIdentifier>
            serial_identifier = soup.find('eop:serialIdentifier').text

            # encontre o valor de <eop:incidenceAngle>
            incidence_angle = soup.find('eop:incidenceAngle').text

            # encontre o valor de <opt:illuminationAzimuthAngle>
            illumination_azimuth_angle = soup.find('opt:illuminationAzimuthAngle').text

            # encontre o valor de <pt:illuminationElevationAngle>
            illumination_elevation_angle = soup.find('opt:illuminationElevationAngle').text

            # encontre o valor de <ps:azimuthAngle>
            azimuth_angle = soup.find('ps:azimuthAngle').text

            # encontre o valor de <ps:spaceCraftViewAngle>
            spacecraft_view_angle = soup.find('ps:spaceCraftViewAngle').text

            # adicionar novas colunas ao dataframe e atribuir os valores extraídos do XML a elas
            df.loc[index, 'serial_identifier'] = serial_identifier
            df.loc[index, 'incidence_angle'] = incidence_angle
            df.loc[index, 'illumination_azimuth_angle'] = illumination_azimuth_angle
            df.loc[index, 'illumination_elevation_angle'] = illumination_elevation_angle
            df.loc[index, 'azimuth_angle'] = azimuth_angle
            df.loc[index, 'spacecraft_view_angle'] = spacecraft_view_angle

            # use o valor de serial_identifier, incidence_angle, illumination_azimuth_angle,
            # illumination_elevation_angle, azimuth_angle e spacecraft_view_angle como necessário
            print(serial_identifier, incidence_angle, illumination_azimuth_angle,
                  illumination_elevation_angle, azimuth_angle, spacecraft_view_angle)
        else:
            print(f"Arquivo XML para '{img_id}' não encontrado.")
    else:
        print(f"Pasta para '{img_id}' não encontrada.")
        
df.to_csv('D:/mestrado/dissertacao/dados/tabelas/resultados/Amz1/DSF_glint_stats_xml.csv', index=False)
```
